# Verification

```python
angle = 45
func = functools.partial(worlds.build_isoscelese_triangle_world, angle)
world = func()
light = definitions.generate_scintillated_light_node(world)
light.translate((0,0,2))
scene = pvtrace.Scene(world)

vis = pvtrace.MeshcatRenderer(wireframe=False, open_browser=True)
vis.render(scene)
```

# Grid trials

```python
world = func()
crystal = world.children[0]

zrange = (
    geometry_helpers.min_physical_value(crystal, 'z')+1.2,
    geometry_helpers.max_physical_value(crystal, 'z')-0.5,
)
yrange = (
    geometry_helpers.min_physical_value(crystal, 'y')+0.1,
    geometry_helpers.max_physical_value(crystal, 'y')-0.1,
)
```

```python
steps = 5
man = processing.Manager(func, out_dir=f'isoscelese_triangle_{angle}deg/')
man.define_grid(
    vertical_range=zrange,
    horizontal_func=lambda z: (-0.3*z-0.01, -(-0.3*z-0.01)),
    steps=steps,
    visualize=True
)
```

## Run

```python
runs = man.run(200, visualize=False)
```

```python
fig, axs = plt.subplots(steps, steps, figsize=(3*steps, 3*steps), layout='constrained', sharex=True, sharey=True)
for i, run in enumerate(runs):
    row = i // steps
    col = i % steps
    ax = axs[row,col]

    events = run.ray_dict
    print('run:', run.index)
    processing.print_event_report(events)
    print('\n\n\n')
    distances = processing.compute_ray_path_lengths(events.all)
    processing.histogram_distances(distances, ax=ax)
    ax.set_title(f'run {run.index}\n{ax.get_title()}')

plt.show()
```

```python
indices = []
surface = []
y, z = [], []
for run in runs:
    distances = processing.compute_ray_path_lengths(run.ray_dict.all)
    y.append(run.light_location[1])
    z.append(run.light_location[2])
    indices.append(run.index)
    print(distances)
    surface.append(np.mean(distances))

Y = np.array(y).reshape((steps,steps))
Z = np.array(z).reshape((steps,steps))
surface = np.array(surface).reshape((steps,steps))

ratio = len(run.ray_dict['sipm_absorbed'])/len(run.ray_dict.all)

fig, ax = plt.subplots()
C = ax.pcolormesh(Y, Z, surface, cmap=plt.cm.coolwarm)
plt.colorbar(C, label='photon path length[mm]')
# ax.set_aspect('equal', adjustable='box')

ax.set(
    # title=f'ratio absorbed by sipm: {ratio:0.3f}',
    title=f'isoscelese triangle {angle}deg 6 sipm',
    # xlim=(-1, 1),
    # ylim=(0, 4)
)
```
